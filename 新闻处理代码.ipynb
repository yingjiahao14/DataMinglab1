```python
import jieba
fstop = open("stopwords/cn_stopwords.txt",'r',encoding='utf-8')
stopwords = []
line = fstop.readline()
while line:
    stopwords.append(line.strip())
    line = fstop.readline()
```

```python
f = open('news.csv', 'r')
line = f.readline()
blockList = ['时政', '国际', '社会', '财经', '产经', '金融', '汽车', '港澳', '台湾', '华人', '娱乐', '体育', '文化']
dirtWordList = [',', '[',']','，','.','。','(',')','?','、','\'',"\"",'‘','’','"','"','（',')','？',' ','“','”','：',
                '！','—','...','《','》']
newsWordDic = dict()
newsWorAnalise = dict()
for blockName in blockList:
    newsWordDic[blockName] = dict()
    newsWorAnalise[blockName] = dict()
while line:
    #print(line)
    line = line.strip().split(",")
    newsTitle = str(line[1:])
    newsWordDic[line[0]][newsTitle] = newsWordDic[line[0]].get(newsTitle, 0) + 1
    line = f.readline()
f.close()
f = open('news1.csv', 'r', encoding='UTF-8') 
line = f.readline()
while line:
    #print(line)
    line = line.strip().split(",")
    newsTitle = str(line[1:])
    newsWordDic[line[0]][newsTitle] = newsWordDic[line[0]].get(newsTitle, 0) + 1
    line = f.readline()
f.close()
for lab in newsWordDic:
    for news in newsWordDic[lab]:
        newsWordList = list(jieba.cut(news))
        for word in newsWordList:
            if word not in dirtWordList and word not in stopwords:
                newsWorAnalise[lab][word] = newsWorAnalise[lab].get(word, 0) + 1
for lab in newsWorAnalise:
    b = sorted(newsWorAnalise[lab].items(), key=lambda x: x[1], reverse=True)
    print(b[0])
```
